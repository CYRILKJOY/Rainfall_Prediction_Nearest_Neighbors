# Rainfall based on the Months Grouped by Year

```python
plt.figure(figsize=(18,10))
rainfall_months = rainfall[['JUN','JUL', 'AUG', 'SEPT',]].groupby('YEAR')
sns.lineplot(data = rainfall_months.mean(),markers = True,hue='YEAR')
plt.title('Rainfall based on the Months and Year')
plt.grid()
plt.show()
```

# Rainfall based on Months

```python
plt.figure(figsize=(18,10))
rainfall_months = rainfall[['JUN','JUL', 'AUG', 'SEPT']].groupby('YEAR')
sns.barplot(data = rainfall_months.mean(),palette = "Blues_d")
sns.set(style="white")
plt.xlabel("Months", fontsize=15)
plt.ylabel("Monthly rainfall", fontsize=15)
plt.grid()
plt.show()
```

# Prediction

```python
rainfall
```

```python
X=rainfall.loc[:,'JUN':'SEPT'].values
X.shape
```

```python
y=rainfall.loc[:,'JUN-SEPT'].values
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test= train_test_split(X,y,test_size = 0.5,random_state=0)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
KNN=KNeighborsRegressor(n_neighbors=5)
```

```python
KNN.fit(X_train,y_train)
```

```python
y_pred=KNN.predict(X_test)
```

```python
y_pred
```

```python
score=KNN.score(X_test,y_test)
```
